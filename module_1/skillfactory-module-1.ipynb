# Предобработка

```python
answers = {} # создадим словарь для ответов

# тут другие ваши предобработки колонок например:

# add profit column to original dataset
data['profit'] = data.revenue - data.budget

# add month of release to original dataset
data['release_month'] = data.release_date.apply(lambda s: s.split('/')[0])

# add title length to original dataset
data['title_length'] = data.original_title.apply(lambda s: len(s))

# add overview words count to original dataset
data['overview_words_count'] = data.overview.apply(lambda s: len(s.split(' ')))

# make dataset with exploded genres
data_genres_exploded = data.copy()
data_genres_exploded['genres'] = data_genres_exploded.genres.apply(lambda s: s.split('|'))
data_genres_exploded = data_genres_exploded.explode('genres')

# make dataset with exploded director (hmm...)
data_director_exploded = data.copy()
data_director_exploded['director'] = data_director_exploded.director.apply(lambda s: s.split('|'))
data_director_exploded = data_director_exploded.explode('director')

# make dataset with director and genres exploded
data_director_genres_exploded = data_director_exploded.copy()
data_director_genres_exploded['genres'] = data_director_genres_exploded.genres.apply(lambda s: s.split('|'))
data_director_genres_exploded = data_director_genres_exploded.explode('genres')

# make dataset with exploded cast
data_cast_exploded = data.copy()
data_cast_exploded['cast2'] = data['cast']
data_cast_exploded['cast'] = data_cast_exploded.cast.apply(lambda s: s.split('|'))
data_cast_exploded = data_cast_exploded.explode('cast')

# make dataset with cast and genres exploded
data_cast_genres_exploded = data_cast_exploded.copy()
data_cast_genres_exploded['genres'] = data_cast_genres_exploded.genres.apply(lambda s: s.split('|'))
data_cast_genres_exploded = data_cast_genres_exploded.explode('genres')

# make dataset with production_companies exploded
data_production_companies_exploded = data.copy()
data_production_companies_exploded['production_companies'] = data_production_companies_exploded.production_companies.apply(lambda s: s.split('|'))
data_production_companies_exploded = data_production_companies_exploded.explode('production_companies')

# make dataset with even more exploded cast
data_cast_exploded['cast2'] = data_cast_exploded.cast2.apply(lambda s: s.split('|'))
data_cast_exploded2 = data_cast_exploded.explode('cast2')
```

```python
data2 = data_cast_exploded2.pivot_table(columns = 'cast', index = 'cast2', values = 'original_title', aggfunc = 'count')
maxVal = 0
maxPair = dict()
i = 1
for cast1, row in data2.items():
    for cast2, val in row.items():
        i += 1
        if cast1 != cast2:
            if val > maxVal:
                maxVal = val
                maxPair[maxVal] = [[cast1, cast2]]
            elif val == maxVal:
                maxPair[maxVal].append([cast1, cast2])

print(maxPair[maxVal])
```
